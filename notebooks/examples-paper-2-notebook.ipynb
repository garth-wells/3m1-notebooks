# MULTIDIMENSIONAL OPTIMISATION

```python
def plot_contour(results, i): # This function plot the contour
    n = 1000
    x = np.linspace(-1.5, 1.5, n)
    y = np.linspace(-1.5, 1.5, n)
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)
    #V = range(0, 1000, 20)
    
    # Create a figure of size 8x6 inches, 80 dots per inch
    pl.figure(figsize=(8, 6), dpi=80)
    CS = pl.contour(X, Y, Z, alpha=1.0, linewidth=.5, cmap = 'spectral')
    # CS = pl.clabel(CS, fontsize=9, inline=1)
    # Plot the direction
    print(" Number of iteration: {}".format(i))
    print(" Variables: {}".format(results[-1]))
    print(" Current function value: {}".format(f(results[-1,0], results[-1,1])))
    pl.plot(results[:,0], results[:,1], '-ro')
    
    return CS
```

```python
def f(x, y): # Rosenbrock's function
    return 100*(y-x**2)**2 + (1-x)**2
```

```python
def df(x,y): # Compute delta operation
    return np.array([400*(x**3)-400*y*x+2*x-2, 200*y-200*(x**2)])
```

```python
def hessian(x,y): # Hessian matrix
    return np.array([[1200*(x**2)-400*y+2, -400*x],[-400*x, 200]])
```

Below is the steepest descent method. 

```python
# Compute the steepest descent method

x = np.array([-1,1])
results = x.T
i = 0
d = -1*df(x[0], x[1])

while np.linalg.norm(d)>=0.005: # Convergence condition
    i = i+1
    d = -1*df(x[0], x[1])
    H = hessian(x[0], x[1])
    alpha = float(d.T.dot(d)/(d.T.dot(H.dot(d))))
    x = x + alpha*d
    results = np.vstack((results,x))
    #print(" Iteration:   {}".format(i))
    #print("     Direction:   {}".format(d))
    #print("     Variables:   {}".format(x))
    #print("     Step size:   {}".format(alpha))


CS = plot_contour(results, i)
```

The Newtons' method below includes a widget that allows user to change how much the variable is moved after each iteration. 

```python
# Compute the Newton's method

def plot_newton(alpha = 0.5):
    x = np.array([-1,1])
    results = x.T # Record all the values of x on an iteration
    i = 0
    e = -1*df(x[0], x[1])

    while np.linalg.norm(e)>=0.005: # Convergence condition
        i = i+1
        e = df(x[0], x[1])
        H = hessian(x[0], x[1])
        H_inv = np.linalg.inv(H) 
        d = -1*H_inv.dot(e)
        x = x + alpha*d
        results = np.vstack((results,x))

    CS = plot_contour(results, i)

# Call the slider interaction
interact(plot_newton, alpha=(0.1,1, 0.1))
```

```python
# Compute the Newton-Raphson method - Still error - alpha remains at 1 as always

x = np.array([-1.0,1.0])
results = x.T # Record all the values of x on an iteration
i = 0
e = -1*df(x[0], x[1])

while np.linalg.norm(e)>=0.005: # Convergence condition
    i = i+1
    e = df(x[0], x[1])
    H = hessian(x[0], x[1])
    H_inv = np.linalg.inv(H) 
    d = -1*H_inv.dot(e)
    alpha = -1.0*float(e.T.dot(d)/(d.T.dot(H.dot(d))))
    x = x + alpha*d
    results = np.vstack((results,x))

CS = plot_contour(results, i)
```

```python
# Compute the Conjugate Gradient Method

x = np.array([-1,1])
results = x.T # Record all the values of x on an iteration
i = 0

# First iteration
d = -1*df(x[0], x[1])
H = hessian(x[0], x[1])
alpha = float(d.T.dot(d)/(d.T.dot(H.dot(d))))
i = 1

while np.linalg.norm(d)>=0.005: # Convergence condition
    x_prev = x
    x = x + alpha*d
    
    # Evaluate next parameters
    df_x_prev = df(x_prev[0], x_prev[1])
    df_x = df(x[0], x[1])
    H = hessian(x[0], x[1])
    
    beta = (np.linalg.norm(df_x)/np.linalg.norm(df_x_prev))**2
    d = -df_x + beta*d
    alpha = -1.0* (d.T.dot(df_x))/(d.T.dot(H.dot(d)))
    
    i = i+1
    results = np.vstack((results,x))

CS = plot_contour(results, i)
```

```python
# Compute the Conjugate Gradient Method - Restart after every two iterations

x = np.array([-1,1])
results = x.T # Record all the values of x on an iteration
i = 0

# First iteration
d = -1.0*df(x[0], x[1])
H = hessian(x[0], x[1])
alpha = float(d.T.dot(d)/(d.T.dot(H.dot(d))))
i = 1

while np.linalg.norm(d)>=0.005: # Convergence condition
    x_prev = x
    x = x + alpha*d
    df_x_prev = df(x_prev[0], x_prev[1])
    df_x = df(x[0], x[1])
    H = hessian(x[0], x[1])
    
    # Evaluate next parameters
    if (i%3) == 0:
        d = -1.0*df_x
        alpha = float(d.T.dot(d)/(d.T.dot(H.dot(d))))
    else:
        beta = (np.linalg.norm(df_x)/np.linalg.norm(df_x_prev))**2
        d = -df_x + beta*d
        alpha = -1.0*float(d.T.dot(df_x))/(d.T.dot(H.dot(d)))
    
    i = i+1
    results = np.vstack((results,x))

CS = plot_contour(results, i)
```
